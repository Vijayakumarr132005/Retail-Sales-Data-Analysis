```python
import pandas as pd
# Load the dataset
# Make sure the CSV file is in the same folder as your Python script/notebook
df = pd.read_csv(r"C:\Users\vijay\OneDrive\Desktop\Resume\new_retail_data.csv")# Take a quick look at the first 5 rows
print(df.head())
```

```python
# Check data types and overall non-null counts
print(df.info())

# Check the exact number of missing values in each column
print(df.isnull().sum())
```

```python
# Drop exact duplicate rows and update the dataframe in place
df.drop_duplicates(inplace=True)

print(f"Rows after dropping duplicates: {len(df)}")
```

```python
# Define the columns that absolutely must have data
critical_columns = ['Transaction_ID', 'Customer_ID', 'Amount', 'Total_Amount', 'Date', 'Product_Category']

# Drop rows missing data in these specific columns
df.dropna(subset=critical_columns, inplace=True)

print(f"Rows after dropping NA: {len(df)}")
```

```python
# 1. Fix IDs: Convert from float to integer, then to string (so they aren't treated as math numbers)
df['Transaction_ID'] = df['Transaction_ID'].astype(int).astype(str)
df['Customer_ID'] = df['Customer_ID'].astype(int).astype(str)

```

```python
# 2. Fix Zipcode: Fill any remaining missing zipcodes with 0, then convert to string
df['Zipcode'] = df['Zipcode'].fillna(0).astype(int).astype(str)
```

```python
# 3. Fix Dates: Convert the Date column to proper datetime objects
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Drop any rows where the Date formatting completely failed
df.dropna(subset=['Date'], inplace=True)

# 4. Clean up Finances: Round monetary columns to 2 decimal places for clean dashboards
df['Amount'] = df['Amount'].round(2)
df['Total_Amount'] = df['Total_Amount'].round(2)
```

```python
# Save to a new CSV file
# index=False prevents pandas from writing row numbers as a new column
df.to_csv(
    r"C:\Users\vijay\OneDrive\Desktop\Resume\cleaned_retail_data.csv",
    index=False
)
print("Data cleaning complete! File saved as cleaned_retail_data.csv")
```
